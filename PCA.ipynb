```python

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data" # load the data
df = pd.read_csv(url
                 , names=['sepal length','sepal width','petal length','petal width','target'])
features  = ['sepal length','sepal width','petal length','petal width',]
```

```python
#seperate the features and  the target
df.columns
y = df['target']
x = df.loc[:,'sepal length':'petal width']

#Transform using the  standard_scaler

x = pd.DataFrame(StandardScaler().fit_transform(x),columns = features)
```

```python
#With the scaled Data, fit a pca model
pca  = PCA(n_components = 2)
pfc = pca.fit_transform(x)
pdf = pd.DataFrame(data = pdf ,columns = ["p-component 1","p-component 2"] )
pdf["Targets"] = y
y.unique()
```

```python
#Graph the data for better visual understanding
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ["green" if i == 'Iris-setosa' else "red" if i == 'Iris-versicolor' else "purple" for i in y]
#colors = ["r","g","b"]

plt.scatter(pdf["p-component 1"],pdf["p-component 2"],color = colors, )

plt.show()

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = pdf['Targets'] == target
    ax.scatter(pdf.loc[indicesToKeep, 'p-component 1']
               , pdf.loc[indicesToKeep, 'p-component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
